**Extracting  text from alldocument and make dataframes**

```python
from bs4 import BeautifulSoup
import pandas as pd
import os

def extract_text_from_doc(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        soup = BeautifulSoup(content, 'lxml')
        text = soup.find('text').get_text()
    return text

root_directory = '/content/TELEGRAPH_UTF8/'

data = []

for root, dirs, files in os.walk(root_directory):
    for file in files:
        if file.endswith('.utf8'):
            file_path = os.path.join(root, file)
            text = extract_text_from_doc(file_path)

            data.append({'docno': file, 'text': text})

df = pd.DataFrame(data)

print(df)
```

```python
df['docno'] = df['docno'].astype(str)
df['text'] = df['text'].astype(str)
```

** Making index of iterdictonary**

```python
from pyterrier.index import IterDictIndexer

docs_as_dict = df.to_dict(orient='records')

indexer = IterDictIndexer("/tmp/index_path")
index_ref = indexer.index(docs_as_dict, meta=['docno', 'text'])
```

**Pre process Step**

```python
import os
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

def preprocess(text):
    tokens = word_tokenize(text.lower())

    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if t not in stop_words]

    stemmer = PorterStemmer()
    tokens = [stemmer.stem(t) for t in tokens]

    return ' '.join(tokens)
```

**Importing nltk**

```python
 import nltk
 nltk.download('punkt')
 nltk.download('stopwords')
```

**Mapping query with qid**

```python
queries = []
with open('/content/FIRE_Dataset_EN_2010/en.topics.76-125.2010.txt', 'r') as file:
    soup = BeautifulSoup(file, "xml")
    for topic in soup.find_all("top"):
        query_num = topic.num.string.strip()
        title = topic.title.string.strip()
        queries.append({'qid': query_num, 'query': preprocess(title)})

q_df = pd.DataFrame(queries)
```

```python
q_df
```

**Finding TF-IDF**

```python

tfidf_retriever = pt.BatchRetrieve(index_ref, controls={"wmodel": "TF_IDF"})

q_df['query'] = q_df['query'].str.replace("'", "")

results = tfidf_retriever.transform(q_df)
```

```python
results
```

```python
qrels = "/content/FIRE_Dataset_EN_2010/en.qrels.76-125.2010.txt.gz"
qrels_df = pt.io.read_qrels(qrels)
```

```python
qrels_df
```

**Evaluateing thoghut Precision of 10 doc , recall and F1 score**

```python
eval_metrics = ["P_10", "recall_10"]
evaluator = pt.Utils.evaluate(results, qrels_df, metrics=eval_metrics)
print(evaluator)
```

```python
if evaluator['P_10'] + evaluator['recall_10'] > 0:
    f1_at_10 = 2 * (evaluator['P_10'] * evaluator['recall_10']) / (evaluator['P_10'] + evaluator['recall_10'])
else:
    f1_at_10 = 0.0

print(f"F1@10: {f1_at_10:.4f}")
```

```python
print("Result of Precision , Recall and F1 Score")
print(evaluator)
print(f"F1@10: {f1_at_10:.4f}")
```
