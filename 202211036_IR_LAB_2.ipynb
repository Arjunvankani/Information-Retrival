**Remove stop words from data**

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def remove_stop_words(word_list):
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in word_list if word.lower() not in stop_words]
    return filtered_words

```

```python
filtered_words = remove_stop_words(all_articles_new)
filtered_words
```

**Stemming sentence data**

```python
from nltk.stem import PorterStemmer

def stem_words(word_list):
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in word_list]
    return stemmed_words


stemmed_words = stem_words(filtered_words)
stemmed_words
```

<h2><b>Q-2) Implement TF-IDF from scratch:

<h4> <b> Calculate term frequency of the words

**Term frequency for Document**

```python
from collections import Counter
import re

def preprocess_text(text):
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
    return cleaned_text.split()

def calculate_tf(documents):
    term_frequencies = []

    for doc in documents:
        words = preprocess_text(doc)
        word_count = Counter(words)
        total_words = len(words)
        tf = {word: count / total_words for word, count in word_count.items()}
        term_frequencies.append(tf)

    return term_frequencies

tf_per_document = calculate_tf(stemmed_words)
for i, tf_dict in enumerate(tf_per_document, start=1):
    print(f"Term frequency for Document {i}:")
    print(tf_dict)
    print()
```

**Inverse Document Frequency:**

```python
import math
import re

def preprocess_text(text):
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
    return set(cleaned_text.split())

def calculate_idf(documents):
    total_documents = len(documents)
    idf = {}

    word_document_count = {}
    for doc in documents:
        words_in_doc = preprocess_text(doc)
        for word in words_in_doc:
            word_document_count[word] = word_document_count.get(word, 0) + 1

    for word, doc_count in word_document_count.items():
        idf[word] = math.log(total_documents / (1 + doc_count))

    return idf



idf = calculate_idf(stemmed_words)
print("Inverse Document Frequency:")
print(idf)
```

**Term-Document Matrix Shape:**

```python
import numpy as np
import re
from collections import Counter

documents = stemmed_words

def preprocess_text(text):
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
    return cleaned_text.split()

vocabulary = set()
for doc in documents:
    words = preprocess_text(doc)
    vocabulary.update(words)

vocabulary = sorted(vocabulary)
num_documents = len(documents)
num_terms = len(vocabulary)


term_document_matrix = np.zeros((num_terms, num_documents), dtype=int)

for doc_idx, doc in enumerate(documents):
    words = preprocess_text(doc)
    word_count = Counter(words)
    for term_idx, term in enumerate(vocabulary):
        term_frequency = word_count.get(term, 0)
        term_document_matrix[term_idx, doc_idx] = term_frequency

document_frequency = np.sum(term_document_matrix > 0, axis=1)
idf = np.log(num_documents / (1 + document_frequency))

tf_idf_matrix = term_document_matrix * idf[:, np.newaxis]

print("Term-Document Matrix Shape:", term_document_matrix.shape)

print("TF-IDF Matrix Shape:", tf_idf_matrix.shape)
```
